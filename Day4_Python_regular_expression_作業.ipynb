### 使用正規表達式對email資料進行處理
這裡我們會使用到python其他的套件協助處理(ex: pandas, email, etc)，這裡我們只需要專注在正規表達式上即可，其他的套件是方便我們整理與處理資料。

### 讀取與切分Email
讀入的email為一個長字串，利用正規表達式切割讀入的資料成一封一封的email，並將結果以list表示。

輸出: [email_1, email_2, email_3, ....]

```python
import re
import pandas as pd
import email

import re
import pandas as pd
import email

###讀取文本資料###
all_emails=open("all_emails.txt",encoding="utf8", errors='ignore').read()


###切割讀入的資料成一封一封的email###
###我們可以使用list來儲存每一封email###
###注意！這裡請仔細觀察sample資料，看資料是如何切分不同email###

# pattern = re.compile(r"From:.*")
# match=re.findall(pattern, all_emails)
# emails = [m for m in match if m]

emails = re.split(r"From ", all_emails, flags=re.M)
emails = emails[1:]
len(emails) #查看有多少封email
```

### 從文本中擷取所有寄件者與收件者的姓名和地址

```python
emails_list = [] #創建空list來儲存所有email資訊

name_pattern = re.compile(r"From:\s*\"([^\"\?\/]*)\"")
address_pattern = re.compile(r"<([^@\-\!\<\>\=]+@[^\<\>]*)>")

for mail in emails[:20]: #只取前20筆資料 (處理速度比較快)
    emails_dict = dict() #創建空字典儲存資訊
    ###取的寄件者姓名與地址###
    
    #Step1: 取的寄件者資訊 (hint: From:)
    sender = re.search(r"From:.*", mail)

    #Step2: 取的姓名與地址 (hint: 要注意有時會有沒取到配對的情況)
    sender_name = re.search(name_pattern, sender.group(0))
    sender_address = re.search(address_pattern, sender.group(0))

    #Step3: 將取得的姓名與地址存入字典中
    emails_dict["sender_name"] = sender_name.group(1) if sender_name else ""
    emails_dict["sender_address"] = sender_address.group(1) if sender_address else ""
        
    
    ###取的收件者姓名與地址###
    #Step1: 取的寄件者資訊 (hint: To:)
    receiver = re.search(r"To:.*", mail)
    
    #Step2: 取的姓名與地址 (hint: 要注意有時會有沒取到配對的情況)
    receiver_name = re.search(name_pattern, sender.group(0))
    receiver_address = re.search(address_pattern, sender.group(0))
        
    #Step3: 將取得的姓名與地址存入字典中
    emails_dict["receiver_name"] = receiver_name.group(1) if sender_name else ""
    emails_dict["receiver_address"] = receiver_address.group(1) if sender_address else ""
        
    ###取得信件日期###
    #Step1: 取得日期資訊 (hint: To:)
    receive_date_meta = re.search(r"Date:.*", mail)

    #Step2: 取得詳細日期(只需取得DD MMM YYYY)
    receive_date = (
        re.search(r"\d{2,}\s\w{3,}\s\d{4,}", receive_date_meta.group(0) ) 
            if receive_date_meta and receive_date_meta.group(0) 
            else ""
    )
    
    #Step3: 將取得的日期資訊存入字典中
    emails_dict["receive_date"] = receive_date     
        
    ###取得信件主旨###
    #Step1: 取得主旨資訊 (hint: Subject:)
    subject_meta = re.search(r"Subject:.*", mail)
    
    #Step2: 移除不必要文字 (hint: Subject: )
    subject = (
        re.sub(r"Subject: ", "", subject_meta.group(0)) 
            if subject_meta and subject_meta.group(0) 
            else ""
    )
    
    #Step3: 將取得的主旨存入字典中
    emails_dict["subject"] = subject
    
    
    ###取得信件內文###
    #這裡我們使用email package來取出email內文 (可以不需深究，本章節重點在正規表達式)
    try:
        full_email = email.message_from_string(mail)
        body = full_email.get_payload()
        emails_dict["email_body"] = body
    except:
        emails_dict["email_body"] = None
    
    ###將字典加入list###
    emails_list.append(emails_dict)
```

```python
#將處理結果轉化為dataframe
emails_df = pd.DataFrame(emails_list)
emails_df
```
